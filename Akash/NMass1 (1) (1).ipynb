```python
q1=df.number_of_bathrooms.quantile(0.25)
q2=df.number_of_bathrooms.quantile(0.75)
```

```python
print(q1)
print(q2)
```

```python
IQR=q2-q1
```

```python
upper_limit=q2+1.5*IQR
lower_limit=q1-1.5*IQR
```

```python
upper_limit
```

```python
df.median()
```

```python
df['number_of_bathrooms']=np.where(df['number_of_bathrooms']>upper_limit,2.25,df['number_of_bathrooms'])
```

```python
sns.boxplot(df.number_of_bathrooms)
```
